```python
import tweepy
import pandas as pd
import re

# authentication
bearer_token = "AAAAAAAAAAAAAAAAAAAAAO3c2gEAAAAAz89TBv8UmPcwcQTHwVec3xQ7jGw%3DkTHDtAW3M8tiZz5dfBWKpRUWKpi5cYVYt0fSj4NYjjZ4MlZ30Y"  # Replace with your actual token
client = tweepy.Client(bearer_token=bearer_token)

# earlier we got non engish data, now i limited to it rnglish tweets only
response = client.search_recent_tweets(
    query="#Agnipath lang:en",
    max_results=50,
    tweet_fields=["created_at", "author_id"],
    expansions=["author_id"],
    user_fields=["username"]
)

# usernames of the person tweeting
users = {u.id: u.username for u in response.includes['users']}

# Prepare data
data = []
for tweet in response.data:
    username = users.get(tweet.author_id, "unknown")
    data.append({
        "original_text": tweet.text,
        "username": username,
        "date": tweet.created_at
    })

df = pd.DataFrame(data)

# cleaning function
def clean_tweet(tweet):
    tweet = tweet.lower()
    tweet = re.sub(r"http\S+|www\S+|https\S+", "", tweet)  # Remove URLs
    tweet = re.sub(r"@\w+", "", tweet)                     # Remove mentions
    tweet = re.sub(r"#\w+", "", tweet)                     # Remove hashtags
    tweet = re.sub(r"[^\w\s]", "", tweet)                  # Remove special chars
    tweet = re.sub(r"\s+", " ", tweet).strip()             # Normalize whitespace
    return tweet

#Applying cleaning function
df["clean_text"] = df["original_text"].apply(clean_tweet)


df = df[["original_text", "clean_text", "username", "date"]]

#Save to CSV
df.to_csv("week2_cleaned.csv", index=False)
print("âœ… Saved cleaned tweets to week2_cleaned.csv")
print(df.head())
```

```python

from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize analyzer
analyzer = SentimentIntensityAnalyzer()

# Sentiment classification function
def get_sentiment_label(score):
    if score >= 0.05: return "Positive"
    elif score <= -0.05: return "Negative"
    else: return "Neutral"

# Calculate sentiment scores
def analyze_sentiment(text):
    # TextBlob analysis
    tb_score = TextBlob(text).sentiment.polarity if isinstance(text, str) else 0.0
    tb_label = get_sentiment_label(tb_score)

    # VADER analysis
    vader_score = analyzer.polarity_scores(text)["compound"] if isinstance(text, str) else 0.0
    vader_label = get_sentiment_label(vader_score)

    return tb_score, tb_label, vader_score, vader_label

# Apply to all tweets
df[['textblob_score', 'textblob_label',
    'vader_score', 'vader_label']] = df['clean_text'].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


df = pd.read_csv("week2_cleaned_budget2025.csv")

# Initialize analyzer
analyzer = SentimentIntensityAnalyzer()

# Sentiment classification function
def get_sentiment_label(score):
    if score >= 0.05: return "Positive"
    elif score <= -0.05: return "Negative"
    else: return "Neutral"

# Calculate sentiment scores
def analyze_sentiment(text):
    # TextBlob analysis
    tb_score = TextBlob(text).sentiment.polarity if isinstance(text, str) else 0.0
    tb_label = get_sentiment_label(tb_score)

    # VADER analysis
    vader_score = analyzer.polarity_scores(text)["compound"] if isinstance(text, str) else 0.0
    vader_label = get_sentiment_label(vader_score)

    return tb_score, tb_label, vader_score, vader_label

# Apply to all tweets
df[['textblob_score', 'textblob_label',
    'vader_score', 'vader_label']] = df['clean_text'].apply(
    lambda x: pd.Series(analyze_sentiment(x))
)

# Save results
df.to_csv('week3_sentiment_budget2025.csv', index=False)
print("âœ… Sentiment analysis complete for #Budget2025 tweets")
print(df[['clean_text', 'textblob_label', 'vader_label']].head())
```
